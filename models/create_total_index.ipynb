```python
rocchio_dict = {}
for rev_ids_lst in rel_dict[" ".join(query_org)]:
    for id in rev_ids_lst:
        for term in newdict[id]:
            rocchio_dict[term] = query.count(term) * 0.8
print(rocchio_dict)
```

```python
frequency = 0
for term in rocchio_dict.keys():
    for i in rel_dict[" ".join(query_org)][0]:
        if term in newdict[i]:
            frequency += 1
    rocchio_dict[term] = rocchio_dict[term] + (abs((frequency/len(rel_dict[" ".join(query_org)][0]))) * 0.3)
    frequency = 0
print(rocchio_dict)
```

```python
frequency = 0
for term in rocchio_dict.keys():
    for i in rel_dict[" ".join(query)][1]:
        if term in newdict[i]:
            frequency += 1
    rocchio_dict[term] = rocchio_dict[term] - ((frequency/len(rel_dict[" ".join(query)][1])) * 0.1)
    frequency = 0
print(rocchio_dict)
```
